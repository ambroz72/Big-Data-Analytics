```python
# Step 1: Import libraries
import cv2
import matplotlib.pyplot as plt

# Step 2: Mount Google Drive (if image is stored there)
from google.colab import drive
drive.mount('/content/drive')

# Step 3: Load the image (update the path as needed)
image_path = '/content/drive/MyDrive/wallpaper (1).jpg'  # <-- Change this!
image = cv2.imread(image_path)

# Step 4: Check if the image loaded successfully
if image is None:
    raise FileNotFoundError("Image not found. Check the path!")

# Step 5: Convert BGR (OpenCV default) to RGB (matplotlib expects RGB)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 6: Display the image using matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.axis('off')  # Optional: Hide axis
plt.title('Displayed Image')
plt.show()
```
